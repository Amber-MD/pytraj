```python
from pytraj import io as mdio
from pytraj import analdict, DataSetList
from pytraj.plots import plot_matrix
```

```python
dslist = DataSetList()
dslist.add_set("coords", "my_rms2d", "")
coords = dslist[0]
coords.load("../tests/data/md1_prod.Tc5b.x", "../tests/data/Tc5b.top")
```

```python
act = analdict['rms2d']
```

```python
act("@CA crdset my_rms2d out ./output/test.out", coords.top, dslist=dslist)
```

```python
print (dslist.get_legends())
```

```python
result = dslist['Rms2D_00001']
print (result)
print (result.get_full_matrix().__len__())
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina' 
import matplotlib
matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']

fig, asp, axi, = plot_matrix(result)
fig.colorbar(axi, ax=asp)
```
