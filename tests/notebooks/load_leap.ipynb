```python
import pytraj as pt

cm = """
source leaprc.protein.ff14SB
TC5b = sequence { NASN LEU TYR ILE GLN TRP LEU LYS
                        ASP GLY GLY PRO SER SER GLY ARG
                        PRO PRO PRO CSER }
saveamberparm TC5b x.parm7 x.crd
"""

traj = pt.io.load_leap(cm)
view = traj.view()
view
```

```python
view.render_image()
```

```python
view._display_image()
```

```python
pt.make_structure(traj, 'alpha:1-12')
```

```python
view.coordinates_dict = {0: traj.xyz[0]}
view.center(0)
view.render_image()
```

```python
view._display_image()
```
